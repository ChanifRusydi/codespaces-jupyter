## Exploratory Data Analysis (EDA)

### Explore and Analyze Data

```python
# plot data correlation
import seaborn as sns
corr = day.corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu')
```

### Convert to category

```python
columns = [
    'season',
    "mnth",
    "holiday",
    "weekday",
    "workingday",
    "weathersit",
]
for column in columns:
    day[column] = day[column].astype('category')
    hour[column] = hour[column].astype('category')
day.info()
hour.info()
```

```python
figure, ax = plt.subplots(figsize=(10, 10))
sns.pointplot(data=hour, x='hr', y='cnt', hue='weekday', ax=ax)
ax.set(title='Bike Sharing Total Users by Hours for each Days in Weekdays', xlabel='Hours', ylabel='Users')
```
